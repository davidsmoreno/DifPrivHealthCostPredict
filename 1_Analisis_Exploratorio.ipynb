### Descripción de Variables del Dataset Médico

- **Código de la EPS (EPS):** Identificador único de la entidad prestadora de salud asignado por el sistema nacional de salud (SNS).

- **Identificación (NUM_IDE):** Número único asignado al afiliado en función del tipo de identificación.

- **Fecha de Nacimiento (FECHA_NACI):** Registro de la fecha de nacimiento del afiliado, con una validación para fechas anteriores al 01/01/1900.

- **Sexo:** Género del afiliado, indicado como Femenino (F) o Masculino (M).

- **Código Departamento (DPTO):** Código asignado al departamento donde reside el afiliado, basado en la división político-administrativa del DANE.

- **Código Municipio (MUNI):** Código del municipio de residencia del afiliado.

- **Días Completados de Tratamiento (DIAS_COMP):** En el contexto del tratamiento médico, podría referirse al número total de días que el paciente ha completado un tratamiento prescrito o ha estado bajo cuidado médico específico, especialmente en casos de tratamientos prolongados como la quimioterapia o rehabilitación.

- **V8:** Solo tiene tres categorías, pero no sabemos en realidad qué significa cada variable.

- **Código del Diagnóstico (Creo que es ID):** Código del diagnóstico principal que motivó la prestación del servicio, basado en la codificación CIE-10.

- **Valor_Neto:** Este sería el costo final que se factura a la entidad pagadora después de aplicar deducciones, seguros, copagos, y cualquier otro tipo de ajuste financiero.

- **DIAS_ESTAN:** El total de días que un paciente ha estado hospitalizado.


- **Edad**: Edad del paciente en años.
- **SIDA_VIH**: Indica si el paciente tiene diagnóstico de SIDA o VIH.
- **TUBERCULOSIS**: Si el paciente ha sido diagnosticado con tuberculosis.
- **CANCER_OTROSCANCER**: Presencia de cualquier tipo de cáncer no especificado en otras categorías.

- **CANCER_MAMA**: Diagnóstico de cáncer de mama.

- **CANCER_GENITALESMASCULINOS**: Cáncer en los órganos genitales masculinos.
- **CANCER_TEJIDOLINFATICOYAFINES**: Cáncer en el tejido linfático y estructuras afines.
- **CANCER_CERVIXINSITU**: Cáncer cervical in situ.
- **AUTOINMUNE**: Presencia de enfermedades autoinmunes.
- **DIABETES**: Indica diagnóstico de diabetes.
- **SINDROMESCONVULSIVOS(EPILEPSIA)**: Presencia de síndromes convulsivos, incluyendo epilepsia.
- **ENFERMEDADCARDIOVASCULAR_OTRA**: Otros tipos de enfermedades cardiovasculares.
- **ENFERMEDADCARDIOVASCULAR_HIPERTENSION**: Presencia de hipertensión arterial.
- **ENFERMEDADPULMONARDELARGADURACION**: Enfermedades pulmonares de larga duración.
- **ASMA**: Diagnóstico de asma.
- **ARTRITISPIOGENASYREACTIVAS**: Artritis piógena y reactiva.
- **ARTRITIS**: Diagnóstico de artritis.
- **ARTROSIS**: Diagnóstico de artrosis.
- **RENAL_OTRA**: Otras enfermedades renales.
- **RENAL_RENALDELARGADURACION**: Enfermedades renales de larga duración.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import shapiro
import statsmodels.api as sm

stat, p_value = shapiro(positive_values['Valor_Neto_Log'] )
print('Shapiro-Wilk Test statistics=%.3f, p-value=%.3f' % (stat, p_value))

if p_value > 0.05:
    print("Sample looks Gaussian (fail to reject H0)")
else:
    print("Sample does not look Gaussian (reject H0)")

plt.figure(figsize=(6,6))
sm.qqplot(positive_values['Valor_Neto_Log'] , line ='45', fit=True)
plt.title('Q-Q plot of Log-Transformed Valor_Neto')
plt.show()
```

Con base en lo que observamos en la gráfica, parece que nuestra distribución no sigue una distribución normal. Sin embargo, utilizaremos la librería disfit, que incluye una función capaz de probar varias distribuciones aleatorias y seleccionar la que mejor se ajuste a nuestros datos.

```python
import pandas as pd
from distfit import distfit

# Assuming 'positive_values' DataFrame already contains the 'Valor_Neto_Log' column

# Initialize distfit
dfit = distfit(todf=True)  # Set todf=True to get the result in a DataFrame format

# Fit the model to the log-transformed 'Valor_Neto' data
dfit.fit_transform(positive_values['Valor_Neto_Log'])

# Plot the results to see which distribution fits best
dfit.plot()
```
